### Further Insights by Date Components

```python
# Average Price by Month
print("\nAverage Price by Month:")
display(df.groupby('month_name')['price'].mean().sort_values(ascending=False))

# Average Number of Reviews by Month
print("\nAverage Number of Reviews by Month:")
display(df.groupby('month_name')['number_of_reviews'].mean().sort_values(ascending=False))

# Average Price by Day of the Week
print("\nAverage Price by Day of the Week:")
display(df.groupby('day_name')['price'].mean().sort_values(ascending=False))

# Average Number of Reviews by Day of the Week
print("\nAverage Number of Reviews by Day of the Week:")
display(df.groupby('day_name')['number_of_reviews'].mean().sort_values(ascending=False))

# Group by Year for overall trends
print("\nListings and Reviews by Year:")
display(df.groupby('year').agg(total_listings=('id', 'count'), average_reviews=('number_of_reviews', 'mean')))
```

```python
import pandas as pd

# Reload the original dataset to restore the 'price' column
df = pd.read_csv('newyork.csv')

# Re-apply cleaning and preprocessing steps
df.dropna(subset=['name', 'host_name', 'last_review', 'reviews_per_month'], inplace=True)
df.columns = df.columns.str.lower().str.strip()
df['last_review'] = pd.to_datetime(df['last_review'], dayfirst=True)

df['name'] = df['name'].str.lower().str.strip()
df['host_name'] = df['host_name'].str.lower().str.strip()
df['neighbourhood_group'] = df['neighbourhood_group'].str.lower().str.strip()
df['neighbourhood'] = df['neighbourhood'].str.lower().str.strip()
df['room_type'] = df['room_type'].str.lower().str.strip()

df['year'] = df['last_review'].dt.year
df['month'] = df['last_review'].dt.month
df['day'] = df['last_review'].dt.day
df['month_name'] = df['last_review'].dt.month_name()
df['day_name'] = df['last_review'].dt.day_name()

# Verify the 'price' column is restored and looks correct
print("Price column description after reloading and cleaning:")
display(df['price'].describe())

print("Re-running price analysis after fixing the data:")
# Average Price by Month
print("\nAverage Price by Month:")
display(df.groupby('month_name')['price'].mean().sort_values(ascending=False))

# Average Price by Day of the Week
print("\nAverage Price by Day of the Week:")
display(df.groupby('day_name')['price'].mean().sort_values(ascending=False))
```
